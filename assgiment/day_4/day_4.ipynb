```python
import pandas as pd
import numpy as np

# Question 1
data = {
    "name": ["Aayush", "Rahul", "Priya", "Neha", "Arjun"],
    "age": [20, 21, 19, 20, 22],
    "marks": [92, 85, 78, 35, 65]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

print("\nFirst 5 Rows")
print(df.head())

print("\nName and Marks")
print(df[["name", "marks"]])

print("\nMarks Greater than 70")
print(df[df["marks"] > 70])

print("\nSorted by Marks")
print(df.sort_values(by="marks", ascending=False))

df["result"] = df["marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")

print("\nWith Result Column")
print(df)

print("\nAverage Marks")
print(df["marks"].mean())

print("\nHighest Marks Student")
print(df[df["marks"] == df["marks"].max()])

print("\nPassed Students")
print((df["marks"] >= 40).sum())

df.set_index("name", inplace=True)

print("\nAfter Setting Name as Index")
print(df)

df["marks"] = df["marks"].fillna(0)

df.rename(columns={"marks": "score"}, inplace=True)

df.drop(columns=["age"], inplace=True)

print("\nFinal DataFrame")
print(df)

df.to_csv("updated_students.csv")

print("\nupdated_students.csv Saved Successfully")
```
